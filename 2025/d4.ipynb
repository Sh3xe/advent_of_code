```python
import numpy as np
from copy import deepcopy
```

```python
lines = open("input4.txt").readlines()
lines = [ [c for c in line.replace("\n", "")] for line in lines]
```

#### Part 1

```python
key_1 = 0

for (i,j) in np.ndindex( (len(lines),len(lines[0])) ):
	# Count the number of adjacent paper rolls
	rolls = 0
	for (di, dj) in np.ndindex((3,3)):
		(di,dj) = (di-1, dj-1)
		if di == 0 and dj == 0:
			continue
		if (i+di) < 0 or (i+di) >= len(lines):
			continue
		if (j+dj) < 0 or (j+dj) >= len(lines[0]):
			continue
		rolls += (lines[i+di][j+dj] == "@")
	
	key_1 += (rolls < 4) and lines[i][j] == "@"

print(f"Key: {key_1}")
```

#### Part 2

```python
def remove_rolls(lines_cpy):
	tot_removed = 0
	for (i,j) in np.ndindex( (len(lines_cpy),len(lines_cpy[0])) ):
		# Count the number of adjacent paper rolls
		rolls = 0
		for (di, dj) in np.ndindex((3,3)):
			(di,dj) = (di-1, dj-1)
			if di == 0 and dj == 0:
				continue
			if (i+di) < 0 or (i+di) >= len(lines_cpy):
				continue
			if (j+dj) < 0 or (j+dj) >= len(lines_cpy[0]):
				continue
			rolls += (lines_cpy[i+di][j+dj] == "@")
		
		if (rolls < 4) and lines_cpy[i][j] == "@":
			lines_cpy[i][j] = "."
			tot_removed += 1

	return tot_removed
```

```python
lines_cpy = deepcopy(lines)
key_2 = 0
while True:
	num_removed = remove_rolls(lines_cpy)
	key_2 += num_removed
	if num_removed == 0:
		break

print(f"Key: {key_2}")
```
