```python
import numpy as np
from copy import deepcopy
import matplotlib.pyplot as plt
```

```python
lines = open("input5.txt").readlines()
lines = [ line.replace("\n", "") for line in lines]

ranges = []
numbers = []
for line in lines:
	if line == "":
		continue

	if "-" in line:
		els = line.split("-")
		ranges.append( (int(els[0]),int(els[1])) )

	else:
		numbers.append(int(line))
```

### Part 1

```python
key_1 = 0
for n in numbers:
	for (lower, upper) in ranges:
		if lower <= n and n <= upper:
			key_1 += 1
			break

print(f"Key: {key_1}")
```

### Part 2

```python
# merge ranges
ranges_cpy1 = []

# Find all unique numbers
unique_nums = np.sort(np.unique(np.array(ranges)))
ranges_lower = unique_nums[:-1]
ranges_upper = unique_nums[1:]

# Find the range that are full
nonempty_ranges = []
for (lower, upper) in zip(ranges_lower, ranges_upper):
	# check that the range is not empty
	mid = (lower + upper) / 2
	for (l,u) in ranges:
		if l <= mid and mid <= u:
			nonempty_ranges.append( (lower, upper) )
			break

# Merge the consecutive ones
merged_ranges = []
cur_l, cur_u = nonempty_ranges[0]
for (l,u) in nonempty_ranges[1:]:
	# merge
	if cur_u == l:
		cur_u = u
	# append when cannot be extended
	else:
		merged_ranges.append( (cur_l, cur_u) )
		cur_l, cur_u = (l,u)

merged_ranges.append((cur_l,cur_u))
```

```python
# Count available ingredients
key_2 = 0
for (l,u) in merged_ranges:
	key_2 += u-l+1

for (l,u) in ranges:
	if l != u:
		continue

	is_included = False
	for (ll,uu) in merged_ranges:
		if ll <= l <= uu:
			is_included = True
	
	key_2 += not is_included

print(f"Key: {key_2}")
```
